```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=5, n_classes=2, random_state=123)
```

```python
sns.scatterplot(data=pd.DataFrame(X), x=0, y=1, hue=y)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
rbf = SVC(kernel='rbf')
```

```python
rbf.fit(X_train, y_train)
```

```python
y_pred = rbf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.1,1,10,100,1000], 'gamma':[1,0.1,0.01,0.001,0.0001], 'kernel':['rbf']}
```

```python
GridSearchCV(SVC(), param_grid, refit=True,cv=5, verbose=3)
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True,cv=5, verbose=3)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```
