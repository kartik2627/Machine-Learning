```python
data['mpg'] = pd.to_numeric(data['mpg'], errors='coerce')
data['wt'] = pd.to_numeric(data['wt'], errors='coerce')
data['hp'] = pd.to_numeric(data['hp'], errors='coerce')
```

```python
sns.scatterplot(data,x = 'hp',y = 'mpg')
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(12,10))
sns.boxplot(data,x='cyl',y='mpg',ax=axes[0,0])
sns.boxplot(data,x='vs',y='mpg',ax=axes[0,1])
sns.scatterplot(data,x='hp',y='mpg',ax=axes[1,0])
sns.scatterplot(data,x='wt',y='mpg',ax=axes[1,1])
plt.show()
```

```python
sns.barplot(data,x='cyl',y='mpg')
plt.show()
```
